```lua
require 'cudnn';
require 'image';
```

### Loading images

```lua
RGB_set = torch.CudaTensor(2,3,224,224)

RGB_set[1] = image.load('trail_images/1_1.jpg')
RGB_set[2] = image.load('trail_images/1_2.jpg')
```

### Loading model

```lua
model = torch.load('resnet-50-template.t7')
model:cuda()
```

### Inference

```lua
RGB_output = model:forward(RGB_set)
print(RGB_output[1])
```

### Changing colormap from RGB to BGR

```lua
BGR_set = torch.CudaTensor(2,3,224,224)
BGR_set[1] = RGB_set[1]:index(1, torch.CudaTensor({3, 2, 1}))
BGR_set[2] = RGB_set[2]:index(1, torch.CudaTensor({3, 2, 1}))
BGR_output = model:forward(BGR_set)
print(BGR_output[1])
```

### Scaling pixel values to range from 0 to 255

```lua
scaled_pixel_set = BGR_set*255
scaled_output = model:forward(scaled_pixel_set)
print(scaled_output[1])
```
